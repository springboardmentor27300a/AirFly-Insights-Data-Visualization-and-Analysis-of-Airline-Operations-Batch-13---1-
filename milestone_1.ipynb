```python
delay_cols = [
    'AIR_SYSTEM_DELAY',
    'SECURITY_DELAY',
    'AIRLINE_DELAY',
    'LATE_AIRCRAFT_DELAY',
    'WEATHER_DELAY'
]

df[delay_cols] = df[delay_cols].fillna(0)
```

```python
df['CANCELLATION_REASON'] = df['CANCELLATION_REASON'].fillna('Not Cancelled')
df["CANCELLATION_REASON"] = df["CANCELLATION_REASON"].replace({
    "A": "Carrier",
    "B": "Weather",
    "C": "NAS",
    "D": "Security"
})

```

```python
df.loc[df['CANCELLED'] == 1, ['DEPARTURE_DELAY','ARRIVAL_DELAY']] = 0
time_columns = [
    'DEPARTURE_TIME', 'ARRIVAL_TIME',
    'TAXI_OUT', 'TAXI_IN',
    'WHEELS_OFF', 'WHEELS_ON',
    'AIR_TIME', 'ELAPSED_TIME'
]

df.loc[df['CANCELLED'] == 1, time_columns] = 0
df['TAIL_NUMBER'] = df['TAIL_NUMBER'].fillna('Unknown')
df['SCHEDULED_TIME'] = df['SCHEDULED_TIME'].fillna(df['SCHEDULED_TIME'].median())
diverted_cols = [
    'ELAPSED_TIME', 'AIR_TIME', 'ARRIVAL_DELAY',
    'WHEELS_ON', 'TAXI_IN', 'ARRIVAL_TIME'
]

df.loc[df['DIVERTED'] == 1, diverted_cols] = 0


```

# Key Performance Indicators (KPIs)

KPIs (Key Performance Indicators) are measurable metrics used to evaluate airline operational performance.

The following KPIs are calculated for this project:

1. Total Flights  
   Total number of flights in the dataset.

2. Cancellation Rate  
   Percentage of flights that were cancelled.

3. Average Arrival Delay  
   Average delay time (in minutes) for arriving flights.

4. Average Departure Delay  
   Average delay time (in minutes) for departing flights.

5. On-Time Performance Rate  
   Percentage of flights arriving within 15 minutes of scheduled arrival time.

6. Delay Cause Distribution  
   Total delay minutes categorized by:
   - Airline Delay
   - Weather Delay
   - Air System (NAS) Delay
   - Security Delay
   - Late Aircraft Delay
