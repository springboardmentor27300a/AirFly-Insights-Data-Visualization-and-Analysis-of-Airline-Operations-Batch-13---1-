```python
data['CANCELLATION_REASON']=data['CANCELLATION_REASON'].replace("A","Airline/Carrier")
data['CANCELLATION_REASON']=data['CANCELLATION_REASON'].replace("B","Weather")
data['CANCELLATION_REASON']=data['CANCELLATION_REASON'].replace("C","National Aviation System")
data['CANCELLATION_REASON']=data['CANCELLATION_REASON'].replace("D","Security")
```

```python
data['CANCELLATION_REASON']=data['CANCELLATION_REASON'].fillna("Not Cancelled")
```

## Format Date time columns

```python
data['FLIGHT_DATE'] = pd.to_datetime(data[['YEAR','MONTH','DAY']])
```

```python
data['FLIGHT_DATE']
```

```python
cols = [
    'SCHEDULED_DEPARTURE',
    'DEPARTURE_TIME',
    'SCHEDULED_ARRIVAL',
    'ARRIVAL_TIME'
]

for c in cols:

    hhmm = data[c].astype('Int64').astype(str).str.zfill(4)

    time_str = hhmm.str[:2] + ':' + hhmm.str[2:]

    data[c + '_DT'] = pd.to_datetime(
        data['FLIGHT_DATE'].astype(str) + ' ' + time_str,
        errors='coerce'
    )
```

## Drop Irrelavent columns

```python
data.drop(['TAIL_NUMBER','FLIGHT_NUMBER','SCHEDULED_DEPARTURE','DEPARTURE_TIME','SCHEDULED_ARRIVAL','ARRIVAL_TIME'],axis=1,inplace=True)
```

## Memory optimization

```python
## Downcast integer columns
int_cols = data.select_dtypes(include=['int64']).columns

for c in int_cols:
    data[c] = pd.to_numeric(data[c], downcast='integer')
```

```python
## Downcast Float columns 
float_cols = data.select_dtypes(include=['float64']).columns

for c in float_cols:
    data[c] = pd.to_numeric(data[c], downcast='float')
```

```python
cat_cols=['CANCELLATION_REASON','ROUTE']
for c in cat_cols:
    if c in data.columns:
        data[c] = data[c].astype('category')
```
