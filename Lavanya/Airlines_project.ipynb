```python
df_oper = df[df['CANCELLED']==0]  # Only flights that actually operated
df_oper['ARRIVAL_DELAY'].describe()
```

```python
month_delay = df_oper.groupby('MONTH')['ARRIVAL_DELAY'].mean()
import matplotlib.pyplot as plt
month_delay.plot(kind='bar', figsize=(8,4))
plt.title("Average Arrival Delay by Month")
plt.xlabel("Month")
plt.ylabel("Average Delay (minutes)")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
df_oper['ARRIVAL_DELAY'].hist(bins=50)
plt.title("Arrival Delay Distribution")
plt.xlabel("Minutes")
plt.ylabel("Number of Flights")
plt.show()
```

```python
# Group by rounded departure delay to reduce noise
avg_arrival_delay = df_oper.groupby(df_oper['DEPARTURE_DELAY']//5*5)['ARRIVAL_DELAY'].mean()

plt.figure(figsize=(10,5))
avg_arrival_delay.plot(kind='line', color='red')
plt.xlabel("Departure Delay (minutes, rounded by 5)")
plt.ylabel("Average Arrival Delay (minutes)")
plt.title("Average Arrival Delay vs Departure Delay (Line Chart)")
plt.grid(True)
plt.show()

print("Type: Bivariate")
print("Variables: DEPARTURE_DELAY, ARRIVAL_DELAY")
print("Observation: Average arrival delay increases steadily with departure delay, showing strong correlation.")
```

```python
airline_delay = df_oper.groupby('AIRLINE')['ARRIVAL_DELAY'].mean().sort_values()

plt.figure(figsize=(10,5))
airline_delay.plot(kind='bar', color='orange')
plt.title("Average Arrival Delay by Airline")
plt.ylabel("Average Delay (minutes)")
plt.show()

print("Type: Bivariate")
print("Variables: AIRLINE, ARRIVAL_DELAY")
print("Observation: Some airlines consistently have higher average delays than others.\n")
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Arrival Delay by Month
plt.figure(figsize=(10,6))
plt.scatter(df_oper['MONTH'], df_oper['ARRIVAL_DELAY'], alpha=0.3, color='purple')
plt.title("Arrival Delay by Month (Scatter Plot)")
plt.xlabel("Month")
plt.ylabel("Arrival Delay (minutes)")
plt.grid(True)
plt.show()

print("Type: Univariate")
print("Variables: MONTH, ARRIVAL_DELAY")
print("Observation: Most flights have delays near 0–20 mins across all months; December shows more extreme delays.")
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Create IS_PEAK column
df['IS_PEAK'] = df['MONTH'].isin([12, 1, 7])

# 2️⃣ Group by peak vs non-peak
peak_stats = df.groupby('IS_PEAK')[['CANCELLED','ARRIVAL_DELAY']].mean()
print(peak_stats)

# 3️⃣ Plot Arrival Delay
plt.figure(figsize=(8,4))
plt.bar(['Non-Peak','Peak'], peak_stats['ARRIVAL_DELAY'], color='orange')
plt.title("Avg Arrival Delay: Peak vs Non-Peak Months")
plt.ylabel("Minutes")
plt.show()

# 4️⃣ Plot Cancellation Rate
plt.figure(figsize=(8,4))
plt.bar(['Non-Peak','Peak'], peak_stats['CANCELLED'], color='skyblue')
plt.title("Avg Cancellation Rate: Peak vs Non-Peak Months")
plt.ylabel("Proportion Cancelled")
plt.show()
```

```python
import plotly.express as px

airport_stats = df.groupby('ORIGIN_AIRPORT').agg({
    'ARRIVAL_DELAY':'mean',
    'FLIGHT_NUMBER':'count'
}).reset_index().sort_values('FLIGHT_NUMBER', ascending=False).head(20)

fig = px.bar(
    airport_stats,
    x='ORIGIN_AIRPORT',
    y='FLIGHT_NUMBER',
    color='ARRIVAL_DELAY',
    color_continuous_scale='Reds',
    labels={'FLIGHT_NUMBER':'Number of Flights', 'ARRIVAL_DELAY':'Avg Arrival Delay (min)'},
    title='Top 20 Airports: Flight Volume & Avg Arrival Delay'
)
fig.show()
```

```python
monthly_cancel = df.groupby('MONTH')['CANCELLED'].mean() * 100
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(monthly_cancel.index, monthly_cancel.values, color='red', s=100)

for i, val in enumerate(monthly_cancel.values):
    plt.text(monthly_cancel.index[i], val+0.5, f"{val:.1f}%", ha='center')

plt.title("Monthly Flight Cancellations (%) - Dot Plot")
plt.xlabel("Month")
plt.ylabel("Cancellation %")
plt.grid(True)
plt.xticks(monthly_cancel.index)
plt.show()
```

```python
cancel_types = df[df['CANCELLED'] == 1]['CANCELLATION_REASON'].value_counts()
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.pie(cancel_types,
        labels=cancel_types.index,
        autopct='%1.1f%%',
        colors=['orange','red','blue','green'],
        startangle=90,
        wedgeprops={'width':0.4})

plt.title("Flight Cancellation Reasons - Donut Chart")
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Example airport coordinates (add more if needed)
coords = {
    'ATL': (33.6407, -84.4277),
    'ORD': (41.9742, -87.9073),
    'DFW': (32.8998, -97.0403),
    'DEN': (39.8561, -104.6737),
    'LAX': (33.9416, -118.4085),
    'SFO': (37.6213, -122.3790)
}

# Count top routes
routes = df.groupby(['ORIGIN_AIRPORT','DESTINATION_AIRPORT']).size().reset_index(name='FLIGHTS')
top_routes = routes.sort_values('FLIGHTS', ascending=False).head(10)

fig = go.Figure()

# Draw route lines
for _, row in top_routes.iterrows():
    o = row['ORIGIN_AIRPORT']
    d = row['DESTINATION_AIRPORT']
    
    if o in coords and d in coords:
        lat = [coords[o][0], coords[d][0]]
        lon = [coords[o][1], coords[d][1]]
        
        fig.add_trace(go.Scattergeo(
            lat=lat,
            lon=lon,
            mode='lines',
            line=dict(width=2),
            opacity=0.6
        ))

# Draw airport markers
for airport,(lat,lon) in coords.items():
    fig.add_trace(go.Scattergeo(
        lat=[lat],
        lon=[lon],
        mode='markers+text',
        text=airport,
        marker=dict(size=10)
    ))

fig.update_layout(
    title="Top Flight Routes Between Major Airports",
    geo=dict(scope='usa')
)

fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select top 10 busiest origin airports
top_airports = df['ORIGIN_AIRPORT'].value_counts().head(10).index
df_top = df[df['ORIGIN_AIRPORT'].isin(top_airports)]

# Create pivot table
heatmap_data = df_top.pivot_table(
    values='ARRIVAL_DELAY',
    index='ORIGIN_AIRPORT',
    columns='DESTINATION_AIRPORT',
    aggfunc='mean'
)

# Replace missing values
heatmap_data = heatmap_data.fillna(0)

# Plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap='coolwarm', linewidths=0.5)

plt.title("Average Arrival Delay Heatmap by Airport Routes")
plt.xlabel("Destination Airport")
plt.ylabel("Origin Airport")
plt.show()
```
