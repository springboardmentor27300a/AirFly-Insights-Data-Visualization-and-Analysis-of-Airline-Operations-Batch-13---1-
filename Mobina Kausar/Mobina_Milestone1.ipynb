```python
import pandas as pd
import numpy as np
```

```python
# Load dataset (low_memory=False avoids dtype warnings)
df = pd.read_csv('/content/drive/MyDrive/flights.csv', low_memory=False)
```

```python
# Store dtypes before optimization
dtypes_before = df.dtypes
```

```python
memory_before = df.memory_usage(deep=True).sum() / 1024**2
print("Memory Before:", memory_before, "MB")

print(df.dtypes.value_counts())
```

```python
# -------- Memory Optimization --------
# Convert object columns with low unique values to 'category'
#This reduces memory usage significantly
for col in df.select_dtypes(include="object").columns:
    if df[col].nunique() < 0.5 * len(df):
        df[col] = df[col].astype("category")
```

```python
# Downcast int64 to smaller integer types (int32/int16)
#This reduces memory usage without losing data
for col in df.select_dtypes(include="int64").columns:
    df[col] = pd.to_numeric(df[col], downcast="integer")
```

```python
# Downcast float64 to float32
#Saves memory, especially useful for large datasets
for col in df.select_dtypes(include="float64").columns:
    df[col] = pd.to_numeric(df[col], downcast="float")
```

```python
memory_after = df.memory_usage(deep=True).sum() / 1024**2

print("Memory After:", memory_after, "MB")
print("Memory Reduced:", memory_before - memory_after, "MB")

print(df.dtypes.value_counts())
```

```python
#stores datatypes after optimization
dtypes_after = df.dtypes
```

```python
# Combine into one dataframe for comparison
comparison = pd.DataFrame({
    "Before": dtypes_before,
    "After": dtypes_after
})

print(comparison[comparison["Before"] != comparison["After"]])
```

```python
# -------- HANDLING MISSING VALUES --------
print("Missing Values BEFORE:")
print(df[["ARRIVAL_DELAY", "DEPARTURE_DELAY", "CANCELLED"]].isnull().sum())
```

```python
# -------- APPLY CLEANING --------
df["ARRIVAL_DELAY"] = df["ARRIVAL_DELAY"].fillna(0)
df["DEPARTURE_DELAY"] = df["DEPARTURE_DELAY"].fillna(0)
df["CANCELLED"] = df["CANCELLED"].fillna(0).astype(int)
```

```python
# -------- CHECK AFTER --------
print("\nMissing Values AFTER:")
print(df[["ARRIVAL_DELAY", "DEPARTURE_DELAY", "CANCELLED"]].isnull().sum())
```

```python
# --------  Format Datetime Column BEFORE --------
print("FLIGHT_DATE exists before?", "FLIGHT_DATE" in df.columns)

# Create proper datetime column
df["FLIGHT_DATE"] = pd.to_datetime(df[["YEAR", "MONTH", "DAY"]])

# -------- AFTER --------
print("FLIGHT_DATE created.")
print(df["FLIGHT_DATE"].head())
```

```python
# --------   DERIVED FEATURES   --------
# Store original raw columns
df["DEPARTURE_TIME_BEFORE"] = df["DEPARTURE_TIME"]
df["MONTH_BEFORE"] = df["MONTH"]

# Create proper date column
df["FLIGHT_DATE"] = pd.to_datetime(df[["YEAR","MONTH","DAY"]])

# Derived features
df["MONTH_AFTER"] = df["FLIGHT_DATE"].dt.month
df["DAY_OF_WEEK"] = df["FLIGHT_DATE"].dt.dayofweek + 1
df["HOUR_AFTER"] = df["DEPARTURE_TIME"] // 100
df["ROUTE"] = df["ORIGIN_AIRPORT"].astype(str) + "-" + df["DESTINATION_AIRPORT"].astype(str)

comparison = df[[
    "DEPARTURE_TIME_BEFORE",
    "HOUR_AFTER",
    "MONTH_BEFORE",
    "MONTH_AFTER",
    "DAY_OF_WEEK",
    "ROUTE"
]].head()

print(comparison)
```

```python
# --------   REMOVE DUPLICATES   --------
rows_before = df.shape[0]

df.drop_duplicates(inplace=True)

rows_after = df.shape[0]

print("Rows Before:", rows_before)
print("Rows After:", rows_after)
print("Duplicates Removed:", rows_before - rows_after)
```

```python
# After all your cleaning steps
df.to_csv('/content/drive/MyDrive/cleaned_flights.csv', index=False)
print("Cleaned dataset saved successfully.")
```
