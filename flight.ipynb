```python
delay_columns = [
    "ARRIVAL_DELAY",
    "DEPARTURE_DELAY",
    "WEATHER_DELAY",
    "AIR_SYSTEM_DELAY",
    "SECURITY_DELAY",
    "AIRLINE_DELAY",
    "LATE_AIRCRAFT_DELAY"
]

for col in delay_columns:
    df[col] = df[col].fillna(0)
```

```python
df.isnull().sum()
```

```python
df["CANCELLATION_REASON"] = df["CANCELLATION_REASON"].fillna("Not Cancelled")
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    df[col] =df[col].fillna(df[col].median())
```

```python
object_cols = df.select_dtypes(include='object').columns

for col in object_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['DEP_HOUR'] = (df['SCHEDULED_DEPARTURE'] // 100)
```

```python
df[["SCHEDULED_DEPARTURE","DEP_HOUR"]].tail()
```

```python
drop_cols = [
    "WHEELS_OFF","WHEELS_ON"]

df = df.drop(columns=drop_cols)
```

```python
df["DATE"] = pd.to_datetime(df[["YEAR","MONTH","DAY"]])
```

```python
df[["YEAR","MONTH","DAY","DATE"]].head()
```

```python
df[["YEAR","MONTH","DAY","DATE"]].tail()
```

```python
df.isnull().sum()
```

```python
df['ROUTE'] = df['ORIGIN_AIRPORT'].astype(str) + "-" + df['DESTINATION_AIRPORT'].astype(str)
```

# Milestone 2: Visual Exploration and Delay Trends 

## Week 3: Univariate Analysis
### 1. Top Airlines by Number of Flights

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

top_airlines = df['AIRLINE'].value_counts()

sns.barplot(x=top_airlines.index, y=top_airlines.values)

plt.title("Top Airlines by Number of Flights")
plt.xlabel("Airline")
plt.ylabel("Flight Count")
plt.xticks(rotation=45)
plt.savefig("top_airlines.png", bbox_inches="tight")
plt.show()
```

### Observations:
- WN operates the highest number of flights.
- DL and AA also handle a large share of total operations.
- Smaller carriers like VX and HA operate significantly fewer flights.
- The distribution shows that a few airlines dominate the market.

### 2. Busiest Months by Flight Volume

```python
plt.figure(figsize=(10,6))

monthly_counts = df['MONTH'].value_counts().sort_index()

plt.plot(monthly_counts.index, monthly_counts.values, marker='o')

plt.title("Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.grid(True)
plt.savefig("busiest_months.png", bbox_inches="tight")
plt.show()
```

### Observations:
- July has the highest number of flights, indicating peak travel season.
- February records the lowest flight count.
- Summer months (June–August) show increased flight operations.
- Overall flight volume remains relatively consistent throughout the year with minor seasonal fluctuations.

### 3. Flight Distribution by Departure Hour

```python
plt.figure(figsize=(12,6))

hour_counts = df['DEP_HOUR'].value_counts().sort_index()

plt.plot(hour_counts.index, hour_counts.values, marker='o')

plt.title("Flight Distribution by Departure Hour")
plt.xlabel("Departure Hour")
plt.ylabel("Number of Flights")

plt.xticks(hour_counts.index)
plt.grid(True)

plt.show()
```

## Week 4: Delay Analysis – Airline and Weather

### 1. Average Delay by Cause

```python
plt.figure(figsize=(10,6))

delay_causes = [
    'AIRLINE_DELAY',
    'WEATHER_DELAY',
    'AIR_SYSTEM_DELAY',
    'LATE_AIRCRAFT_DELAY',
    'SECURITY_DELAY'
]

avg_delay = df[delay_causes].mean()

sns.barplot(x=avg_delay.index, y=avg_delay.values)

plt.title("Average Delay by Cause")
plt.xlabel("Delay Cause")
plt.ylabel("Average Delay (Minutes)")
plt.xticks(rotation=45)
plt.savefig("average_delay_by_cause.png", bbox_inches="tight")
plt.show()
```

### Observations:
- Late aircraft delay is the largest contributor to overall arrival delays.
- Airline-related delays are also a significant factor.
- Air system (NAS) delays moderately impact flight punctuality.
- Weather delays contribute less on average compared to operational delays.
- Security delays have minimal overall impact.

### 2. Delay Causes by Airline

```python
plt.figure(figsize=(12,6))

airline_delay = df.groupby('AIRLINE')[[
    'AIRLINE_DELAY',
    'WEATHER_DELAY',
    'AIR_SYSTEM_DELAY',
    'LATE_AIRCRAFT_DELAY'
]].mean()

airline_delay.plot(kind='bar', figsize=(14,6))

plt.title("Average Delay Causes by Airline")
plt.xlabel("Airline")
plt.ylabel("Average Delay (Minutes)")
plt.xticks(rotation=45)

plt.show()
```

### Observations:
- Late aircraft delay is a major contributor across most airlines.
- Airline-related delays are consistently present but vary slightly by carrier.
- NK and F9 show higher air system (NAS) delays compared to others.
- Weather delays have relatively minimal impact across airlines.
- Operational and aircraft rotation issues appear to contribute more than weather.

### 3. Average Arrival Delay by Departure Hour

```python
plt.figure(figsize=(12,6))

hour_delay = df.groupby('DEP_HOUR')['ARRIVAL_DELAY'].mean()

plt.plot(hour_delay.index, hour_delay.values, marker='o')

plt.title("Average Arrival Delay by Departure Hour")
plt.xlabel("Departure Hour")
plt.ylabel("Average Arrival Delay (Minutes)")
plt.grid(True)
plt.savefig("avg_delay_by_hour.png", bbox_inches="tight")
plt.show()
```

### Observations:
- Early morning departures experience minimal or even negative average delays.
- Average delays increase progressively throughout the day.
- Evening flights (6 PM – 8 PM) show the highest average arrival delays.
- The peak delay occurs around 7 PM.
- Delays tend to reduce slightly during late-night hours.

### 4. Average Arrival Delay by Airport

```python
plt.figure(figsize=(12,6))

airport_delay = (
    df.groupby('DESTINATION_AIRPORT')['ARRIVAL_DELAY']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

sns.barplot(x=airport_delay.index, y=airport_delay.values)

plt.title("Top 10 Destination Airports with Highest Average Arrival Delay")
plt.xlabel("Destination Airport")
plt.ylabel("Average Arrival Delay (Minutes)")
plt.xticks(rotation=45)

plt.show()
```

### Observations:
- Destination airport 10728 records the highest average arrival delay.
- Airport 14588 and 13264 also show significantly high delay levels.
- There is a noticeable gap between the top 3 airports and the remaining airports.
- Delay severity varies across airports, indicating location-specific operational challenges.
- Congestion, air traffic control constraints, and infrastructure limitations may contribute to higher delays at certain airports.

**Note:** Airport codes shown in the chart represent FAA airport identifiers provided in the dataset.

### 3. Delay Heatmap by Airport and Departure Hour

```python
# Select top airports first
top_airports = df['ORIGIN_AIRPORT'].value_counts().head(10).index

# Create pivot table
heatmap_data = df[df['ORIGIN_AIRPORT'].isin(top_airports)].pivot_table(
    values='ARRIVAL_DELAY',
    index='ORIGIN_AIRPORT',
    columns='DEP_HOUR',
    aggfunc='mean'
)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="Reds")

plt.title("Average Arrival Delay by Airport and Departure Hour")
plt.xlabel("Departure Hour")
plt.ylabel("Origin Airport")

plt.show()
```

### Observations

- The heatmap shows how **average arrival delays vary across airports and different departure hours**.
- **Evening hours (around 17:00–21:00)** generally show higher delay values across multiple airports.
- Some airports such as **DEN, ORD, and LAX** experience relatively higher delays during peak hours, possibly due to congestion and heavy traffic.
- Early morning hours tend to have **lower delay levels**, indicating smoother operations before daily traffic builds up.
- The pattern suggests a **delay accumulation effect**, where delays increase as the day progresses due to operational and scheduling pressures.

### 4. Route Congestion vs Average Delay

```python
route_stats = df.groupby('ROUTE').agg(
    flight_count=('ROUTE','count'),
    avg_delay=('ARRIVAL_DELAY','mean')
).reset_index()

top_routes = route_stats.sort_values('flight_count', ascending=False).head(20)

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=top_routes,
    x='flight_count',
    y='avg_delay',
    s=120
)

plt.title("Route Congestion vs Average Delay")
plt.xlabel("Number of Flights (Route Traffic)")
plt.ylabel("Average Arrival Delay (Minutes)")
plt.savefig("route_congestion_vs_delay.png", bbox_inches="tight")
plt.show()
```

### Observations

- The scatter plot compares **route traffic volume (number of flights)** with **average arrival delay**.
- Some highly congested routes with large flight volumes show **moderate to high delays**, suggesting that traffic congestion can influence operational delays.
- However, not all busy routes experience high delays, indicating that **efficient airport operations and airline management can reduce delays even on busy routes**.
- A few routes with lower traffic still show noticeable delays, implying that **factors such as weather, air traffic control constraints, or scheduling issues may also affect delays**.
- Overall, the relationship between route congestion and delays is **not strictly linear**, highlighting that multiple operational factors influence flight punctuality.

### Objective
To visualize the geographical distribution of major airports and analyze how flight traffic and delays vary across locations.

```python
import plotly.express as px

# Airport statistics
airport_stats = df.groupby("ORIGIN_AIRPORT").agg(
    Flights=("ORIGIN_AIRPORT", "count"),
    Avg_Delay=("ARRIVAL_DELAY", "mean")
).reset_index()

# Select top airports
top_airports = airport_stats.sort_values("Flights", ascending=False).head(15)

# Coordinates of major airports
coords = {
'ATL':[33.64,-84.42],'LAX':[33.94,-118.40],'ORD':[41.97,-87.90],
'DFW':[32.89,-97.04],'DEN':[39.85,-104.67],'JFK':[40.64,-73.77],
'SFO':[37.62,-122.37],'SEA':[47.45,-122.30],'LAS':[36.08,-115.15],
'MCO':[28.43,-81.30],'EWR':[40.68,-74.17],'CLT':[35.21,-80.94],
'PHX':[33.43,-112.01],'IAH':[29.99,-95.33],'MIA':[25.79,-80.28]
}

# Map latitude and longitude
top_airports["lat"] = top_airports["ORIGIN_AIRPORT"].map(lambda x: coords.get(x,[None,None])[0])
top_airports["lon"] = top_airports["ORIGIN_AIRPORT"].map(lambda x: coords.get(x,[None,None])[1])

# Create map
fig = px.scatter_geo(
    top_airports,
    lat="lat",
    lon="lon",
    size="Flights",
    color="Avg_Delay",
    hover_name="ORIGIN_AIRPORT",
    projection="natural earth",
    color_continuous_scale="Plasma",
    title="Map of Busiest Airports and Average Arrival Delays"
)
fig.write_html("airport_delay_map.html")
fig.show()
```

### Observations
- The map shows the geographical locations of major airports along with their flight traffic and average delays.
- Larger bubbles represent airports with higher flight volumes, indicating major airline hubs.
- Color intensity represents the average delay, where darker colors indicate higher delay values.
- Airports such as ATL, ORD, and DFW appear as major traffic hubs with significant flight activity.
- The visualization helps understand how airport location and traffic concentration influence delay patterns.

## Week 6: Seasonal and Cancellation Analysis 

### 1. Monthly Cancellation Trend

```python
cancel_month = df[df['CANCELLED'] == 1]['MONTH'].value_counts().sort_index()

plt.figure(figsize=(10,6))
sns.lineplot(x=cancel_month.index, y=cancel_month.values, marker='o')

plt.title("Monthly Flight Cancellation Trend")
plt.xlabel("Month")
plt.ylabel("Number of Cancellations")
plt.savefig("monthly_cancellations.png", bbox_inches="tight")
plt.show()
```

### Observations

- The highest number of flight cancellations occurs in **February**, indicating a peak cancellation period.
- **January and March** also show relatively high cancellation counts compared to other months.
- A noticeable decline in cancellations is observed from **April to May**, suggesting more stable flight operations during this period.
- Cancellations increase again slightly around **June and December**, which may be associated with seasonal travel demand and weather conditions.
- The lowest cancellation levels are observed around **September and October**, indicating comparatively smoother flight operations during these months.

### 2. Flight Cancellation Reasons

```python
cancelled_flights = df[df['CANCELLED'] == 1]

cancel_reason = cancelled_flights['CANCELLATION_REASON'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=cancel_reason.index, y=cancel_reason.values)

plt.title("Distribution of Flight Cancellation Reasons")
plt.xlabel("Cancellation Reason")
plt.ylabel("Number of Cancelled Flights")
plt.savefig("cancellation_reasons.png", bbox_inches="tight")
plt.show()
```

### Observations

- **Weather-related cancellations (B)** appear to be one of the most common reasons for flight cancellations.
- **Airline-related issues (A)** also contribute significantly to cancelled flights.
- **NAS (National Airspace System) related cancellations (C)** represent delays caused by air traffic control and system congestion.
- **Security-related cancellations (D)** occur very rarely compared to other causes.
- The results indicate that **weather and operational factors are the dominant causes of flight cancellations**.

### 3. Winter Cancellation Analysis

```python
winter_months = df[df['MONTH'].isin([12, 1, 2])]

winter_cancel = winter_months['CANCELLED'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=winter_cancel.index, y=winter_cancel.values)

plt.title("Flight Cancellations During Winter Months")
plt.xlabel("Cancelled (0 = No, 1 = Yes)")
plt.ylabel("Number of Flights")
plt.savefig("winter_cancellations.png", bbox_inches="tight")
plt.show()
```

### Observations

- The majority of flights during winter months are **not cancelled**, indicating that airline operations remain largely stable even during colder seasons.
- However, a noticeable number of flights are still **cancelled during winter**, which may be influenced by adverse weather conditions such as snow, storms, or reduced visibility.

### 4. Heatmap of Delay Causes by Airline

```python
import seaborn as sns
import matplotlib.pyplot as plt

delay_data = df.groupby("AIRLINE")[[
    "AIRLINE_DELAY",
    "WEATHER_DELAY",
    "AIR_SYSTEM_DELAY",
    "LATE_AIRCRAFT_DELAY"
]].mean()

plt.figure(figsize=(10,6))
sns.heatmap(delay_data, annot=True, cmap="coolwarm")

plt.title("Heatmap of Average Delay Causes by Airline")
plt.savefig("delay_cause_heatmap.png", bbox_inches="tight")
plt.show()
```

- **Air System Delay** is particularly high for some airlines such as **NK (Spirit)**, suggesting that factors like air traffic congestion, airport operations, or system-related issues impact these flights more.
- **Airline Delay** values remain moderately consistent across most airlines, reflecting operational or internal airline-related delays.
- **Weather Delay** is relatively low across nearly all airlines, indicating that weather contributes less to average delay time compared to other operational factors.
- Overall, the heatmap helps identify **patterns of delay causes across airlines**, showing that operational and aircraft-related factors tend to contribute more to delays than weather conditions.
